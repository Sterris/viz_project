### Short description of daily sets

We have a total of three datasets about covid-19.
- Cases per day
- Deaths per day
- Recovered per day

Each of the datasets report the number daily cases from *22. jan to 6. april*   for 208 contries worldwide. Thus the datasets are 208 rows x 76 columns. Source: https://data.humdata.org/dataset/novel-coronavirus-2019-ncov-cases


```python
#Simple plot of total cases by day
x_ticks = []
i=0
for col in df:
    if(i%5 == 0):
        x_ticks.append(col)
    i = i +1


plt.plot(df.columns, df.sum())
plt.xticks(x_ticks, rotation = 45)
plt.xlabel("Date")
plt.ylabel("Total cases")
plt.title("Covid-19 overview")
plt.show()
```
